```python
df = pd.read_csv("before_scaling_employee.csv")
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
```

```python
#extract numerical columns for scaling
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
```

```python
standard_scaler = StandardScaler()
df_standard_scaled = df.copy()
df_standard_scaled[numerical_columns] = standard_scaler.fit_transform(df[numerical_columns])
```

```python
minmax_scaler = MinMaxScaler()
df_minmax_scaled = df.copy()
df_minmax_scaled[numerical_columns] = minmax_scaler.fit_transform(df[numerical_columns])
```

```python
df_standard_scaled.to_csv("standard_scaled_employee.csv", index=False)
df_minmax_scaled.to_csv("minmax_scaled_employee.csv", index=False)
```
